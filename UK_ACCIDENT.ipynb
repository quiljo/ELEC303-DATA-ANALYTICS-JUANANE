```python
# 🧹 Clean up 'District Area' column to avoid matching issues
uk['District Area'] = uk['District Area'].str.strip().str.lower()

# Step 1: Filter only Fatal accidents with valid lat/lon
fatal_df = uk[uk['Accident_Severity'] == 'Fatal'].dropna(subset=['Latitude', 'Longitude'])

# Step 2: Separate data for each area
london_df = fatal_df[fatal_df['District Area'].str.contains('south northamptonshire')]
dudley_df = fatal_df[fatal_df['District Area'] == 'salisbury']

# Count fatalities for each area
london_count = len(london_df)
dudley_count = len(dudley_df)

# Step 3: Create base map centered roughly between the two
map_center = [fatal_df['Latitude'].mean(), fatal_df['Longitude'].mean()]
m = folium.Map(location=map_center, zoom_start=7)

# Step 4: Create FeatureGroups for each city with counts in the layer name
london_layer = folium.FeatureGroup(name=f'South Northamptonshire Fatal Accidents ({london_count})')
dudley_layer = folium.FeatureGroup(name=f'Salisbury Fatal Accidents ({dudley_count})')

# Step 5: Add heatmaps to layers
HeatMap(list(zip(london_df['Latitude'], london_df['Longitude']))).add_to(london_layer)
HeatMap(list(zip(dudley_df['Latitude'], dudley_df['Longitude']))).add_to(dudley_layer)

# Step 6: Add layers to map
london_layer.add_to(m)
dudley_layer.add_to(m)

# Step 7: Add layer control toggle
folium.LayerControl().add_to(m)

# Step 8: Save and/or display
m.save('layered_fatal_heatmap.html')
m  # Show in Jupyter if applicable
```

<h1>INSIGHT 31</h1>
<h1>Between South Northamptonshire and Salisbury, District Salisbury have the most fatal accidents with 31 that have ever recorded while the South Northamptonshire with only 28 Fatal accidents.  </h1>
<hr>






















```python

# Step 1: Standardize 'District Area'
uk['District Area'] = uk['District Area'].str.strip().str.lower()

# Step 2: Filter fatal accidents with valid coordinates
fatal_df = uk[uk['Accident_Severity'] == 'Fatal'].dropna(subset=['Latitude', 'Longitude'])

# Step 3: Count fatal accidents per district
fatal_counts = fatal_df['District Area'].value_counts()

# Step 4: Keep only areas with >60 fatal accidents
high_fatal_areas = fatal_counts[fatal_counts > 60].index

# Step 5: Base map centered on all fatal accidents
map_center = [fatal_df['Latitude'].mean(), fatal_df['Longitude'].mean()]
m = folium.Map(location=map_center, zoom_start=6)

# Step 6: Add a heatmap layer per high-fatality area
for area in high_fatal_areas:
    area_df = fatal_df[fatal_df['District Area'] == area]
    
    # Create layer with proper title case
    heat_layer = folium.FeatureGroup(name=f"{area.title()} ({len(area_df)} fatalities)")
    
    # Add heatmap
    HeatMap(list(zip(area_df['Latitude'], area_df['Longitude']))).add_to(heat_layer)
    
    # Add to map
    heat_layer.add_to(m)

# Step 7: Add layer control toggle
folium.LayerControl(collapsed=False).add_to(m)

# Step 8: Save or display
m.save('high_fatal_areas_heatmap.html')
m  # Display in Jupyter if applicable
```
